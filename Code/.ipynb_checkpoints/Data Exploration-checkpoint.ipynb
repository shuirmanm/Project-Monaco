# Analysis - Getting a feel for the data

## Analysis - Who was the winner of the Drivers' Championship the past 10 years?

```python
# Merging the race results with the year
RaceYear = results.merge(races, on=['raceId'], how='left')
RaceSprintYear = RaceYear.merge(sprint_results, on=['raceId','driverId'], how='left')
RaceSprintYear['points'] = RaceSprintYear['points_x'] + RaceSprintYear['points_y'].fillna(0)
RaceSprintYear.head()
```

```python
AnnualPoints = RaceSprintYear.groupby(['driverId','year'], as_index=False).points.sum()
AnnualPoints.head()
```

```python
Winners = AnnualPoints.loc[AnnualPoints.groupby("year")["points"].idxmax()]
WinnersLastDecade = Winners.loc[((Winners['year'] >= 2012) & (Winners['year'] < 2022))]
Names = drivers[['driverId','surname']]
WinnersNames = WinnersLastDecade.merge(Names, on=['driverId'], how='left')
print(WinnersNames)
```

## Analysis - Who won the Constructors' Championship the past 10 years?

```python
AnnualPointsC = RaceSprintYear.groupby(['constructorId_x','year'], as_index=False).points.sum()
```

```python
WinnersC = AnnualPointsC.loc[AnnualPointsC.groupby("year")["points"].idxmax()]
WinnersLastDecadeC = WinnersC.loc[((WinnersC['year'] >= 2012) & (WinnersC['year'] < 2022))]
WinnersLastDecadeC2 = WinnersLastDecadeC.rename(columns={'constructorId_x':'constructorId'})
NamesC = constructors[['constructorId','name']]
WinnersNamesC = WinnersLastDecadeC2.merge(NamesC, on=['constructorId'], how='left')
print(WinnersNamesC)
```

# Sample ML Model

```python
y = results.points

results_features = ['driverId','constructorId','number','grid']
X = results[results_features]
X.loc[X['number'] == '\\N','number'] = 0
```

```python
from sklearn.tree import DecisionTreeRegressor

sample_model = DecisionTreeRegressor(random_state=1)

sample_model.fit(X, y)
```

```python

results2 = results
results2['prediction'] = sample_model.predict(X)
results2.head()
```
