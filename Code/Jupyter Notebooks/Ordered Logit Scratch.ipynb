```python
#putting together some ugly code for now, will clean up after
class DeepLearningClassifier(object):
    def __init__(self):
        
        num_classes = 20
        
        self.model = tf.keras.Sequential()
        self.model.add(tf.keras.layers.Dense(128, activation = "relu"))
        self.model.add(tf.keras.layers.Dropout(0.2))
        self.model.add(tf.keras.layers.Dense(32, activation = "relu"))
        self.model.add(tf.keras.layers.Dropout(0.1))
      # No activation function specified so this will output cumulative logits.
        self.model.add(tf.keras.layers.Dense(num_classes-1))       

        self.model.compile(loss = condor.SparseCondorOrdinalCrossEntropy(), optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))

    
    def train(self, training_data, num_epochs):
  
    
        #This code should probably be deleted and implemented upstream
        training_data["weather_warm"] = training_data["weather_warm"].astype(int)
        training_data["weather_cold"] = training_data["weather_cold"].astype(int)
        training_data["weather_dry"] = training_data["weather_dry"].astype(int)
        training_data["weather_wet"] = training_data["weather_wet"].astype(int)
        training_data["weather_cloudy"] = training_data["weather_cloudy"].astype(int)

        x = training_data.drop(columns = ['podium','driver_points_after_race.1','driver_wins_after_race.1','driver_standings_pos_after_race.1'])
        y = training_data['podium']

        
        x = np.asarray(x).astype('float32')
        y = np.asarray(y).astype('float32')

        mean = tf.math.reduce_mean(x, axis=0)
        variance = tf.math.reduce_variance(x, axis=0)

        # Create the Normalization layer
        self.normalization_layer = tf.keras.layers.experimental.preprocessing.Normalization(mean=mean, variance=variance)
        
        #x_normalized = self.normalization_layer(x)
        
        #Takes any race outcome greater than 
        #I don't like the fact that we need this line 
        #We will definitely want this to work differently when we add DNFs as a separate class
        for i in range(len(y)):
            if y[i] > 20:
                y[i] = 20
       
        #This line is to get values so they range from 0-19 instead of 1-20
        #might need to delete this line - I am not sure how it works with the other code
        y = y-1 
        

        self.model.fit(x,y,epochs=num_epochs)
        
    def test(self, testing_data):
        
        
        testing_data["weather_warm"] = testing_data["weather_warm"].astype(int)
        testing_data["weather_cold"] = testing_data["weather_cold"].astype(int)
        testing_data["weather_dry"] = testing_data["weather_dry"].astype(int)
        testing_data["weather_wet"] = testing_data["weather_wet"].astype(int)
        testing_data["weather_cloudy"] = testing_data["weather_cloudy"].astype(int)     
        
        x_test = testing_data.drop(columns = ['podium','driver_points_after_race.1','driver_wins_after_race.1','driver_standings_pos_after_race.1'])
        y_test = testing_data['podium']
        
        
        x_test = np.asarray(x_test).astype('float32')
        y_test = np.asarray(y_test).astype('float32')
 
        #x_test_normalized = self.normalization_layer(x_test)

        for i in range(len(y_test)):
            if y_test[i] > 20:
                y_test[i] = 20
               
        y_test = y_test-1 
                
        test_loss = self.model.evaluate(x_test, y_test)
        
        ordinal_logits = self.model.predict(x_test)

        # Convert from logits to label probabilities. This is initially a tensorflow tensor.
        preds = condor.ordinal_softmax(ordinal_logits)
        
        return test_loss, preds
    
    def reset(self):
        pass
```
