```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
pd.set_option('display.max_columns', None)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
```

```python
df = pd.read_csv('../Race Data/final_df.csv',header = 0,sep = ',')
df.head()
```

```python
#removing an unnecessary column
df.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
#removing columns that shouldn't be in the dataset 
df.drop(columns = ['grid', 'driver_points_after_race.1','driver_wins_after_race.1','driver_standings_pos_after_race.1','qualifying_time'], inplace = True)
```

```python
df['season'].min()
```

Test 1 - basic neural net

```python
#establishing x and y
x = df.drop(columns = 'podium')
y = df['podium']

x = np.asarray(x).astype('float32')
y = np.asarray(y).astype('float32')

```

```python
tf.random.set_seed(1234) # for consistent results
model = Sequential(
    [               
       
        Dense(units = 100, activation ='relu'),
        Dense(units = 50, activation ='relu'),
        Dense(units = 28, activation ='linear')
        

    ]
)
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
              , optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))
model.fit(x,y,epochs=50)
```

```python
preds = model.predict(x)
```

```python
preds_df = pd.DataFrame(tf.nn.softmax(preds))
preds_df.head()
```

Data investigation

```python
df = pd.read_csv('../Race Data/final_df.csv',header = 0,sep = ',')
df.head()
```

```python
df.loc[df['podium'] == 0]
```

```python
df.loc[df['podium'] > 20,['podium']] = 20
```

```python
x = df[['grid','driver_points','driver_wins']]
y = df['podium']

x = np.asarray(x).astype('float32')
y = np.asarray(y).astype('float32')

y = y-1
```

```python
np.unique(y)
```

```python
tf.random.set_seed(1234) # for consistent results
model = Sequential(
    [               
       
        Dense(units = 10, activation ='relu'),
        Dense(units = 20, activation ='linear')
        

    ]
)
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
              , optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))
model.fit(x,y,epochs=20)
```

```python
preds = model.predict(x)
preds_df = pd.DataFrame(tf.nn.softmax(preds))
preds_df.columns += 1
preds_df.head(100)
```

```python

overview = pd.concat([df, preds_df.reindex(df.index)], axis= 1)
overview.head()
```
