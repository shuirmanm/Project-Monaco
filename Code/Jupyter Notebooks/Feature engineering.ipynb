```python
results = pd.read_csv('../../Processed Data/Race Data/results.csv')
```

```python
results.loc[(results['season'] == 2022) & (results['round'] == 2)]
```

```python
results['driver_points_cumulative'] = results.groupby(['driver', 'season'])['points'].cumsum()
results['constructor_points_cumulative'] = results.groupby(['constructor', 'season'])['points'].cumsum()

newdf = pd.DataFrame(results.groupby(['season','round'], as_index=False)['points'].max())
newdf['max_points'] = newdf['points']
newdf.head()
```

```python
new_results = pd.merge(results, newdf, how="left", on=['season','round'], suffixes=('', '_y'))

new_results.drop(new_results.filter(regex='_y$').columns, axis=1, inplace=True)

new_results['max_points_cumulative'] = new_results.groupby(['season','driver'])['max_points'].cumsum()

#Simply doubling the max points - this is technically wrong but should suffice mathematically
new_results['constructor_max_points_cumulative'] = new_results['max_points_cumulative'] * 2

new_results['driver_points_percent_max'] = new_results['driver_points_cumulative'] / new_results['max_points_cumulative']
 
new_results['constructor_points_percent_max'] = new_results['constructor_points_cumulative'] / new_results['constructor_max_points_cumulative']    
    
new_results['round'] = new_results['round'] + 1    
    
```

```python
new_results.loc[(new_results['season'] == 2022) & (new_results['round'] == 22)]
```

```python
new_results_2 = pd.merge(results, new_results, how="left", on=['season','round','driver'], suffixes=('', '_y'))

new_results_2.drop(new_results_2.filter(regex='_y$').columns, axis=1, inplace=True)

new_results_2.drop(columns = ['max_points','max_points_cumulative'], inplace=True)

new_results_2.loc[new_results_2['round'] == 1,['driver_points_percent_max']] = 0
new_results_2.loc[new_results_2['round'] == 1,['constructor_points_percent_max']] = 0

new_results_2['driver_points_cumulative'] = new_results_2['driver_points_cumulative'] - new_results_2['points']

#The constructor points part isn't working and probably needs to be completely reworked
new_results_2.drop(columns = ['constructor_points_cumulative', 'constructor_max_points_cumulative', 'constructor_points_percent_max','grid','date_of_birth','nationality','circuit_id','constructor','podium','points'], inplace=True)

new_results_2.head()
```
